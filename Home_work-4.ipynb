```python
import pandas as pd
titanic=pd.read_csv("train.csv")
titanic.head()
```

```python
def ages(row):
    if row['Age'] < 25:
        return "before_25"
    elif row['Age'] >= 25 and row['Age'] < 50 :
        return "before_50"
    else :
        return "older_than_50"
titanic['ages'] = titanic.apply (lambda row: ages(row), axis=1)
pd.crosstab(index=titanic['ages'], columns=titanic['Survived']).stack()/len(titanic)
```
